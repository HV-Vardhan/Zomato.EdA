## __Libraries used__
* __Numpy__ is used for mathematical or numerical computation, it works on Numpy can work on multidimensional data, which makes it suitable for every dataset available.
* __Pandas__ - This library is built on top of NumPy and provides high-level data structures and tools for data analysis. > `What can we do with the missing values?`
> - Yes in  most of the cases when missing values are low we can impute them with mean of value of that particular column
> - If the percent of missing values is greater than 50% we certianly drop the whole column in that case
> - missing value column is Numeric?  Impute null values with mean and median
> - missing value column is categorical? Impute null values with mode 
> - **Since Cuisines is an obeject type we cannot impute it so we will simply remove all the rows with null values**

```python
# removing the missing values
zomato.dropna(subset='Cuisines', inplace=True)
```

```python
# checking the null values again to verify 
zomato['Cuisines'].isnull().sum()

# so we have no null values currently 
```

> Analysing the Aggregate rating column  

```python
rating=zomato_final.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:'Rating_count'})
```

`OBSERVATION` : We can observe that the unrated values are quite high that is 2148 
further we can see the distribution of the ratings 

__Ranges__
- Not rated : 0
- Poor : 1 - 2.4
- Average : 2.5 - 3.4
- Good : 3.5 - 3.9
- Very good : 4.0 - 4.4
- Excellent : 4.5 - 5.0

>Binnining the values for better visualization 

```python
# Define bin edges
bins = [0, 0.1, 2.4, 3.4, 3.9, 4.4, 5.0]

# Define bin labels
labels = ['Not rated', 'Poor', 'Average', 'Good', 'Very good', 'Excellent']
```

```python
# Apply binning
zomato_final['rating_category'] = pd.cut(zomato_final['Aggregate rating'], bins=bins, labels=labels, include_lowest=True)

# Display the DataFrame with the binned values
print(zomato_final)
```

visualizing using charts

```python

# Plot the bar chart
colors = ['gray', 'red', 'orange', 'yellow', 'green', 'blue']
zomato_final['rating_category'].value_counts().sort_index().plot(kind='bar', color=colors, edgecolor='black')
# Add labels and title
plt.xlabel('Rating Category')
plt.ylabel('Frequency')
plt.title('Frequency of Rating Categories')

# Display the bar chart
plt.show()
```

```python
# Filter the DataFrame for rows where the "Rating text" is "Average"
filtered_df = zomato_final[zomato_final["Rating text"] == "Average"].value_counts().sum()
filtered_df
```
